```python
import pandas as pd
import numpy as np

df = pd.read_csv('10개토픽모델링버전.csv')
table = pd.read_csv('10개토픽테이블.csv')
```

```python
table.drop('Unnamed: 0',inplace=True, axis=1)
df.drop('Unnamed: 0',inplace=True, axis=1)
```

```python
table['각 토픽의 비중'][0]
```

- 홀수는 int로, 짝수는 float로

```python
for i in range(len(df)):
    string = table['각 토픽의 비중'][i]
    print(type(string))
    string = string.replace('[','')
    string = string.replace(']','')
    string = string.replace('(','')
    string = string.replace(')','')
    splits = string.split(", ")
    
    splits = str(splits)
    splits = splits.replace('[','')
    splits = splits.replace(']','')
    splits = splits.replace('\'','')
    table['각 토픽의 비중'][i] = splits
```

```python
# 1. 1~5번 등장비중을 비교하고, 가장 높은 거의 등장토픽을 확인
# 2. 이를 토픽 번호와 비
lists=[]
# 토픽번호하고 5번등장토픽 비교해서 토픽번호가 더 작은거
tos = ['1번등장토픽','2번등장토픽','3번등장토픽','4번등장토픽','5번등장토픽']
for i in range(len(final)):
    if final['토픽 번호'][i] != final['1번등장토픽'][i]:
        if final['토픽 번호'][i] != final[tos[1]][i]:
            if final['토픽 번호'][i] != final[tos[2]][i]:
                if final['토픽 번호'][i] != final[tos[3]][i]:
                    if final['토픽 번호'][i] != final[tos[4]][i]:
                        if final['토픽 번호'][i] != final['6번등장토픽'][i]:
                            lists.append(i)
```

```python
highest_rate = []
for i in range(len(final)):
    if final['토픽 번호'][i] == final['1번등장토픽'][i]:
        highest_rate.append(final['1번등장비중'][i])
    elif final['토픽 번호'][i] == final['2번등장토픽'][i]:
        highest_rate.append(final['2번등장비중'][i])
    elif final['토픽 번호'][i] == final['3번등장토픽'][i]:
        highest_rate.append(final['3번등장비중'][i])
    elif final['토픽 번호'][i] == final['4번등장토픽'][i]:
        highest_rate.append(final['4번등장비중'][i])
    elif final['토픽 번호'][i] == final['5번등장토픽'][i]:
        highest_rate.append(final['5번등장비중'][i])
    elif final['토픽 번호'][i] == final['6번등장토픽'][i]:
        highest_rate.append(final['6번등장비중'][i])
    elif final['토픽 번호'][i] == final['7번등장토픽'][i]:
        highest_rate.append(final['7번등장비중'][i])
    elif final['토픽 번호'][i] == final['8번등장토픽'][i]:
        highest_rate.append(final['8번등장비중'][i])
    elif final['토픽 번호'][i] == final['9번등장토픽'][i]:
        highest_rate.append(final['9번등장비중'][i])
    else:
        highest_rate.append(final['10번등장비중'][i])
        
```
